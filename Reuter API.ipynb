```python
import requests
```

```python
url = ('https://newsapi.org/v2/everything?'
       'sources=cnbc&'
       'apiKey=22238ce39ef4436d8b4bf3796090c6c0')
response = requests.get(url)
print(response.json())
```

```python
articles = response.json()['articles']
print("Total number of articles: {}.\n".format(len(articles)))
for article in articles:
    # each article is a dict
    print("======================================================")
    for key in article:
        print("{key} : {val}\n".format(key = key, val = article.get(key, 'NA')))
```
